# Analisando os dados por local de votação no mapa

```python
# import libraries

import folium
import pandas as pd
```

```python
df = pd.read_csv('dataset/detalhe_votacao_localizacao_consolidado_duque_de_caxias.csv')

dfl = df.sort_values(by='abstencoes', ascending=False)
```

## 30 locais com mais abstenções

```python
dc = folium.Map(
    location=[-22.73619, -43.28884], # come from wego.here.com,
    zoom_start = 11
)

abstencoes_rows = dfl.head(30).to_dict(orient='index')

for local in abstencoes_rows.values():
    abstencoes = local['abstencoes']
    _local = local['nome_local']
    number = (abstencoes / local['aptos']) * 100
    percent = round(number, 2)
    folium.Marker(
        location=[local['lat'], local['lon']],
        tooltip = '{} - {} abstencoes ({}%)'.format(_local, abstencoes, percent)
    ).add_to(dc)
    
dc
```

## 30 locais com mais votos nulos

```python
dfl = df.sort_values(by='votos_nulos', ascending=False)

nulos_rows = dfl.head(30).to_dict(orient='index')

dc = folium.Map(
    location=[-22.73619, -43.28884], # come from wego.here.com,
    zoom_start = 11
)

for local in nulos_rows.values():
    abstencoes = local['votos_nulos']
    _local = local['nome_local']
    number = (abstencoes / local['aptos']) * 100
    percent = round(number, 2)
    folium.Marker(
        location=[local['lat'], local['lon']],
        tooltip = '{} - {} Nulos ({}%)'.format(_local, abstencoes, percent)
    ).add_to(dc)
    
dc
```
